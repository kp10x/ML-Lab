# K Means Difference

```python
import matplotlib
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
import sklearn.metrics as sm
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("heart.csv")
data.sample(10)
```

```python
X = data[['age', 'sex', 'chol', 'thalach', 'fbs', 'oldpeak']]
y = data['heartdisease']
```

```python
model = KMeans(n_clusters=3)
model.fit(X)
```

```python
print('The accuracy score of K-Mean: ', sm.accuracy_score(y, model.labels_))
print('The Confusion matrixof K-Mean: ', sm.confusion_matrix(y, model.labels_))
```

```python
plt.figure(figsize=(20, 10))
plt.subplot(1, 2, 2)
colormap = np.array(['#e74c3c', '#2ecc71', '#9b59b6', '#3498db', '#f1c40f', '#e67e22', '#34495e'])
plt.scatter(X.age, X.chol, c=colormap[model.labels_], s=40)
plt.title('K Mean Classification')
plt.xlabel('age')
plt.ylabel('chol')
```

```python
plt.figure(figsize=(20, 10))
plt.subplot(1, 2, 2)
colormap = np.array(['#e74c3c', '#2ecc71', '#9b59b6', '#3498db', '#f1c40f', '#e67e22', '#34495e'])
plt.scatter(X.chol, X.thalach, c=colormap[model.labels_], s=40)
plt.title('K Mean Classification')
plt.xlabel('chol')
plt.ylabel('thalach')
```
